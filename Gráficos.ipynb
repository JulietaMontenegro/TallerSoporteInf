Hacemos un gráfico de barras para saber si tenemos más hombres o mujeres registradas

```python
import matplotlib.pyplot as plt1
import pandas as pd
import numpy as np

csv = pd.read_csv("./nuevo.csv")
csvColumnas = list(csv.columns)

mujeres = (csv['Gender'] == 'f').sum()
hombres = (csv['Gender'] == 'm').sum()

plt1.figure(figsize=(2, 8))
plt1.bar(['Hombres', 'Mujeres'], [hombres, mujeres], color='blue')
plt1.bar_label(plt1.bar(["Hombres", "Mujeres"], [hombres, mujeres]))
plt1.title('Hombres y mujeres registrados')
plt1.show()

```

Conclusión:Tenemos más mujeres que hombres registrados.

Hacemos un gráfico circular para saber el porcentaje de suscripción de cada tipo disponible

```python
import matplotlib.pyplot as plt1
import pandas as pd
import numpy as np

csv = pd.read_csv("./nuevo.csv")
csvColumnas = list(csv.columns)

suscripciones = csv['Subscription Type'].value_counts()
labels = suscripciones.index
sizes = suscripciones.values
explode = (0.2, 0.1, 0.1)
colores= ['pink', 'purple', 'blue']
plt1.pie(sizes, labels=labels, explode=explode, colors=colores, autopct='%.2f%%',startangle=120, pctdistance=0.5)
plt1.title('Suscripciones de Netflix')
plt1.show()
```

Conclusión: La suscripción con mayores usuarios es la básica con el 39.87%, la segunda la standard con el 30.72% y por último la premium con el 29.32%.

```python
import pandas as pd
import matplotlib.pyplot as plt1


csv = pd.read_csv('./nuevo.csv')
csvColumnas = list(csv.columns)

rangos = [18, 30, 40, 50, 60, 70, float('inf')]

plt1.figure(figsize=(10, 6))
rangosCant = pd.cut(csv['Age'], bins=rangos, right=False).value_counts().sort_index()
plt1.plot(rangosCant.index.categories.left, rangosCant.values, color="red")

plt1.xlabel('Edades')
plt1.ylabel('Cantidad de usuarios')
plt1.title('Usuarios por edades')

plt1.show()

```
